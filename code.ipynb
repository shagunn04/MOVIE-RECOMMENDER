```python
import ast  
```

```python
def convert(obj):
    L=[];
    for i in ast.literal_eval(obj) :
      L.append(i["name"])
    return L;
```

```python
movies["genres"]=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies["keywords"]=movies['keywords'].apply(convert)
```

```python


def convertCast(obj):
    L = []
    count = 0 
    for i in ast.literal_eval(obj):
        if count > 6:  
            break
        L.append(i["name"])  
        count += 1  
    return L
```

```python
movies["cast"]=movies['cast'].apply(convertCast)
```

```python
movies.head(1)
```

```python


def convertCrew(obj):
    L = [] 
    
    for i in ast.literal_eval(obj):
       
        if i["job"] == "Director" or i["job"] == "ScreenPlay" or i["job"] == "Original Music Composer":
            L.append(i["name"]) 
        
    return L
```

```python
movies["crew"]=movies["crew"].apply(convertCrew);
```

```python
movies.head()
```

```python
movies["overview"]=movies["overview"].apply(lambda x : x.split());
```

```python
movies["genres"]=movies["genres"].apply(lambda x:[i.replace(" ","") for i in x ]);
movies["keywords"]=movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x ]);
movies["crew"]=movies["crew"].apply(lambda x:[i.replace(" ","") for i in x ]);
movies["cast"]=movies["cast"].apply(lambda x:[i.replace(" ","") for i in x ]);
```

```python
movies["tags"]=movies["genres"]+movies["keywords"]+movies["crew"]+movies["cast"]+movies["overview"]
```

```python
def recommend(movie):
    L = []
    try:
        movie_index = new_df[new_df["title"] == movie].index[0]
    except IndexError:
        return ["Movie not found in database."]
    movie_vector = vectors[movie_index].reshape(1, -1)
    distances, indices = knn.kneighbors(movie_vector, n_neighbors=6) 
    for i in range(1, len(indices[0])):  
        L.append(new_df.iloc[indices[0][i]].title)
    
    return L
```

```python
def hit_rate_at_k(model_df, test_df, k=7):
    hits = 0
    total = 0
    for title in test_df['title']:
        if title not in model_df['title'].values:
            continue
        recommended = recommend(title)
        if any(movie in recommended for movie in test_df['title'].values):
            hits += 1
        total += 1

    return hits / total if total > 0 else 0
    
accuracy = hit_rate_at_k(train_df, test_df, k=7)
print(f"Hit Rate@5 Accuracy: {accuracy:.2f}")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

def compute_confusion_matrix(model_df, test_df, k=5):
    y_true = []
    y_pred = []

    for title in test_df['title']:
        if title not in model_df['title'].values:
            continue

        recommended = recommend(title)

        y_true.append(1)  
        y_pred.append(1 if any(movie in recommended for movie in test_df['title'].values) else 0)

    cm = confusion_matrix(y_true, y_pred, labels=[1, 0])
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Hit", "Miss"])
    disp.plot()
    return cm
```

```python
cm = compute_confusion_matrix(train_df, test_df, k=5)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(new_df['tags'])
print("TF-IDF matrix shape:", tfidf_matrix.shape)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

best_score = -1
best_k = 0
sil_scores = []

K = list(range(2, 30))  
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(tfidf_matrix)
    labels = kmeans.labels_
    score = silhouette_score(tfidf_matrix, labels, metric='cosine')
    sil_scores.append(score)

    if score > best_score:
        best_score = score
        best_k = k

print(f"Best k: {best_k} with Silhouette Score: {best_score:.4f}")

plt.figure(figsize=(8, 5))
plt.plot(K, sil_scores, marker='o', color='orange')
plt.title("Silhouette Score vs Number of Clusters")
plt.xlabel("k")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=17, random_state=42)
kmeans.fit(tfidf_matrix)
new_df['cluster'] = kmeans.labels_
print(new_df[['title', 'cluster']])
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
pca = PCA(n_components=2)
pca_result = pca.fit_transform(tfidf_matrix.toarray())
plt.figure(figsize=(10,6))
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=new_df['cluster'], cmap='tab20')
plt.title("Movie Clusters (PCA Visualization)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.colorbar(label='Cluster')
plt.show()
```
